```python
from gurobipy import *
import pandas as pd
import numpy as np
import os

# display settings
from IPython.display import display
pd.options.display.max_columns = None
pd.options.display.max_rows = None
pd.set_option('display.float_format', lambda x: '%.4f' % x)
```

## data input

```python
print(os.getcwd())
os.chdir("../to students/data")
print(os.getcwd())
os.listdir()
```

```python
instance = pd.read_csv("instances 1.csv")
display(instance)
```

```python
process_time = instance["processing_time"]
due_time = instance["due_time"]
job_id = instance["job_ID"]
sub_id = instance["subjob"]
boil_dummy = instance["boiling"]
machine = [1,2,3,4,5]
L = sum(process_time) * 100
print(L)
```

## Modeling

### part1

```python
p1 = Model("p1")
j_num = int(len(job_id) / 2)
m_len = len(machine)

# Variables
s_jkm = []
c_jkm = []
x_jkm = []
for j in range(j_num):
    tmp_s = []
    tmp_c = []
    tmp_x = []
    for k in range(2):
        tmp_s.append([])
        tmp_c.append([])
        tmp_x.append([])
        for m in machine:
            tmp_s[k].append(p1.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "s_" + str(j + 1) + "," + str(k + 1) + "," + str(m)))
            tmp_c[k].append(p1.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "c_" + str(j + 1) + "," + str(k + 1) + "," + str(m)))
            tmp_x[k].append(p1.addVar(lb = 0, vtype = GRB.BINARY, name = "x_" + str(j + 1) + "," + str(k) + ","+ str(m)))  
    s_jkm.append(tmp_s)
    c_jkm.append(tmp_c)
    x_jkm.append(tmp_x)

y_jkjkm = []
for j1 in range(j_num):
    y_jkjkm.append([])
    for k1 in range(2):
        y_jkjkm[j1].append([])
        for j2 in range(j1 + 1, j_num):
            y_jkjkm[j1][k1].append([])
            for k2 in range(2):
                y_jkjkm[j1][k1][j2 - j1 - 1].append([])
                for m in machine:
#                     print(j1 + 1, k1 + 1, j2 + 1, k2 + 1, m)
                    y_jkjkm[j1][k1][j2 - j1 - 1][k2].append(p1.addVar(lb = 0, vtype = GRB.BINARY, 
                                                   name = "y_" + str(j1 + 1) + "," + str(k1 + 1) + "," + str(j2 + 1) + "," + str(k2 + 1)+ "," + str(m)))

c_j = []
for j in range(j_num):
    c_j.append(p1.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "c_" + str(j + 1)))
t_j = []
for j in range(j_num):
    t_j.append(p1.addVar(lb = 0, vtype = GRB.BINARY, name = "t_" + str(j + 1)))
```

```python
# setting the objective function 
p1.setObjective(
    quicksum(t_j[j] for j in range(j_num))
    , GRB.MINIMIZE) 
```

```python
# add constraints and name them
# boil
p1.addConstrs((x_jkm[j][k][0] <= boil_dummy[j * 2 + k] for j in range(j_num) for k in range(2)))
p1.addConstrs((t_j[j] * L >= c_j[j] - due_time[2 * j] 
                for j in range(j_num)), "tardy count")
p1.addConstrs((c_j[j] >= quicksum(c_jkm[j][1][m] for m in range(m_len))
                for j in range(j_num)), "job completion time")
p1.addConstrs((quicksum(x_jkm[j][k][m] for m in range(m_len)) == 1 for j in range(j_num) for k in range(2)), "job assignment constraint")
p1.addConstrs((s_jkm[j][k][m] + c_jkm[j][k][m] <= x_jkm[j][k][m] * L for j in range(j_num) for k in range(2) for m in range(m_len)))
p1.addConstrs((c_jkm[j][k][m] >= s_jkm[j][k][m] + process_time[2 * j + k] - (1 - x_jkm[j][k][m]) * L for j in range(j_num) for k in range(2) for m in range(m_len)))
for m in range(m_len):
    for j1 in range(j_num):
        for k1 in range(2):
            for j2 in range(j_num - j1 - 1):
                for k2 in range(2):
                # 注意ijm的indexing方法不同
    #             print(i, j, m)
    #             print(i, i + j + 1, m)
    #             print()
                    p1.addConstr(s_jkm[j1][k1][m] >= c_jkm[j1 + j2 + 1][k2][m] - y_jkjkm[j1][k1][j2][k2][m] * L)
                    p1.addConstr(s_jkm[j1 + j2 + 1][k2][m] >= c_jkm[j1][k1][m] - (1 - y_jkjkm[j1][k1][j2][k2][m]) * L)
p1.addConstrs((quicksum(s_jkm[j][1][m] for m in range(m_len)) >= quicksum(c_jkm[j][0][m] for m in range(m_len)) for j in range(j_num)), "subjob 1 ends before subjob2 starts")
```

```python
p1.optimize()
```

```python
print("Result:") 
for j in range(j_num):
    for k in range(2):
        for m in range(m_len):
            if x_jkm[j][k][m].x != 0:
                print(x_jkm[j][k][m].varName, "=", x_jkm[j][k][m].x)
```

### part2

```python
p1_2 = Model("p1_2")
p1_2.setParam('TimeLimit', 5*60)

# tardy job num result from part 1
T = 0

# Variables
u = p1_2.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "u")
s_jkm = []
c_jkm = []
x_jkm = []
for j in range(j_num):
    tmp_s = []
    tmp_c = []
    tmp_x = []
    for k in range(2):
        tmp_s.append([])
        tmp_c.append([])
        tmp_x.append([])
        for m in machine:
            tmp_s[k].append(p1_2.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "s_" + str(j + 1) + "," + str(k + 1) + "," + str(m)))
            tmp_c[k].append(p1_2.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "c_" + str(j + 1) + "," + str(k + 1) + "," + str(m)))
            tmp_x[k].append(p1_2.addVar(lb = 0, vtype = GRB.BINARY, name = "x_" + str(j + 1) + "," + str(k) + ","+ str(m)))  
    s_jkm.append(tmp_s)
    c_jkm.append(tmp_c)
    x_jkm.append(tmp_x)

y_jkjkm = []
for j1 in range(j_num):
    y_jkjkm.append([])
    for k1 in range(2):
        y_jkjkm[j1].append([])
        for j2 in range(j1 + 1, j_num):
            y_jkjkm[j1][k1].append([])
            for k2 in range(2):
                y_jkjkm[j1][k1][j2 - j1 - 1].append([])
                for m in machine:
#                     print(j1 + 1, k1 + 1, j2 + 1, k2 + 1, m)
                    y_jkjkm[j1][k1][j2 - j1 - 1][k2].append(p1_2.addVar(lb = 0, vtype = GRB.BINARY, 
                                                   name = "y_" + str(j1 + 1) + "," + str(k1 + 1) + "," + str(j2 + 1) + "," + str(k2 + 1)+ "," + str(m)))

c_j = []
for j in range(j_num):
    c_j.append(p1_2.addVar(lb = 0, vtype = GRB.CONTINUOUS, name = "c_" + str(j + 1)))
t_j = []
for j in range(j_num):
    t_j.append(p1_2.addVar(lb = 0, vtype = GRB.BINARY, name = "t_" + str(j + 1)))
```

```python
# setting the objective function 
p1_2.setObjective(u, GRB.MINIMIZE) 
```

```python
# new constraints
p1_2.addConstrs((x_jkm[j][k][0] <= boil_dummy[j * 2 + k] for j in range(j_num) for k in range(2)))
p1_2.addConstrs((u >= c_j[j] for j in range(j_num)))
p1_2.addConstr((quicksum(t_j[j] for j in range(j_num)) == T))

# add constraints and name them
p1_2.addConstrs((t_j[j] * L >= c_j[j] - due_time[2 * j] 
                for j in range(j_num)), "tardy count")
p1_2.addConstrs((c_j[j] >= quicksum(c_jkm[j][1][m] for m in range(m_len))
                for j in range(j_num)), "job completion time")
p1_2.addConstrs((quicksum(x_jkm[j][k][m] for m in range(m_len)) == 1 
               for j in range(j_num) for k in range(2)), "job assignment constraint")
p1_2.addConstrs((s_jkm[j][k][m] + c_jkm[j][k][m] <= x_jkm[j][k][m] * L 
               for j in range(j_num) for k in range(2) for m in range(m_len)))
p1_2.addConstrs((c_jkm[j][k][m] >= s_jkm[j][k][m] + process_time[2 * j + k] - (1 - x_jkm[j][k][m]) * L 
               for j in range(j_num) for k in range(2) for m in range(m_len)))
for m in range(m_len):
    for j1 in range(j_num):
        for k1 in range(2):
            for j2 in range(j_num - j1 - 1):
                for k2 in range(2):
                # 注意ijm的indexing方法不同
    #             print(i, j, m)
    #             print(i, i + j + 1, m)
    #             print()
                    p1_2.addConstr(s_jkm[j1][k1][m] >= c_jkm[j1 + j2 + 1][k2][m] - y_jkjkm[j1][k1][j2][k2][m] * L)
                    p1_2.addConstr(s_jkm[j1 + j2 + 1][k2][m] >= c_jkm[j1][k1][m] - (1 - y_jkjkm[j1][k1][j2][k2][m]) * L)
p1_2.addConstrs((quicksum(s_jkm[j][1][m] for m in range(m_len)) >= quicksum(c_jkm[j][0][m] for m in range(m_len)) for j in range(j_num)), "subjob 1 ends before subjob2 starts")
```

```python
p1_2.optimize()
```

```python
print("Result:") 
for j in range(j_num):
    for k in range(2):
        for m in range(m_len):
            if x_jkm[j][k][m].x != 0:
                print(x_jkm[j][k][m].varName, "=", x_jkm[j][k][m].x)
```

```python
job_schedule = []
for j in range(j_num):
    for k in range(2):
        for m in range(m_len):
            if c_jkm[j][k][m].x > 0:
                job_schedule.append([m + 1, j + 1, k + 1, s_jkm[j][k][m].x, c_jkm[j][k][m].x])
#                 print(s_jkm[j][k][m].varname, "=", s_jkm[j][k][m].x)
#                 print(c_jkm[j][k][m].varname, "=", c_jkm[j][k][m].x)
job_schedule.sort(key=lambda x: (x[0], x[3]))
job_schedule = np.around(np.array(job_schedule), 2)

schedule_df = pd.DataFrame(job_schedule, columns=["machine", "job_ID", "subjob", "start_time", "end_time"])
schedule_df["job_ID"] = schedule_df["job_ID"].astype(int)
schedule_df["subjob"] = schedule_df["subjob"].astype(int)
schedule_df["machine"] = schedule_df["machine"].astype(int)
display(schedule_df)
```
